```python
import cv2
import numpy as np
from ultralytics import YOLO

# تحميل النموذج
model = YOLO("yolov8n.pt")

# تحميل صورة من لقطة عرقلة
image_path = "foul_1.jpg"
output_path = "result_foul_1.jpg"
image = cv2.imread(image_path)

# تحليل الصورة
results = model(image)

players = []
ball = None

# استخراج اللاعبين والكرة
for result in results:
    for box in result.boxes:
        cls_id = int(box.cls[0])
        label = result.names[cls_id]
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        if label in ["person"]:
            center = ((x1 + x2) // 2, (y1 + y2) // 2)
            players.append((x1, y1, x2, y2, center))
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 255), 2)
        elif label in ["sports ball", "ball"]:
            ball = (x1, y1, x2, y2)
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

# التحقق من اقتراب لاعبين أكثر من اللازم = اشتباه عرقلة
def detect_possible_foul(players, threshold=60):
    fouls = []
    for i in range(len(players)):
        for j in range(i+1, len(players)):
            c1 = players[i][4]
            c2 = players[j][4]
            dist = np.linalg.norm(np.array(c1) - np.array(c2))
            if dist < threshold:
                fouls.append((players[i], players[j]))
    return fouls

# كشف العرقلة
foul_pairs = detect_possible_foul(players)

# رسم النتائج
for p1, p2 in foul_pairs:
    x1, y1, x2, y2, _ = p1
    x3, y3, x4, y4, _ = p2
    cv2.putText(image, "Foul?", (min(x1,x3), min(y1,y3)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,0,255), 2)
    cv2.rectangle(image, (x1, y1), (x2, y2), (255, 0, 0), 2)
    cv2.rectangle(image, (x3, y3), (x4, y4), (255, 0, 0), 2)

# حفظ الصورة
cv2.imwrite(output_path, image)
print("تم حفظ الصورة بعد كشف العرقلة المحتملة.")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO

# تحميل نموذج YOLO
model = YOLO("yolov8n.pt")  # تأكد أن لديك النموذج داخل مجلد المشروع

# مسار الصورة
image_path = "foul_1.jpg"       # غيّر هذا للمسار الفعلي للصورة
output_path = "foul_marked_1.jpg"

# قراءة الصورة
image = cv2.imread(image_path)

# تحليل الصورة
results = model(image)
players = []
ball = None

# استخراج اللاعبين والكرة
for result in results:
    for box in result.boxes:
        cls_id = int(box.cls[0])
        label = result.names[cls_id]
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        center = ((x1 + x2) // 2, (y1 + y2) // 2)

        if label == "person":
            players.append((x1, y1, x2, y2, center))
            cv2.rectangle(image, (x1, y1), (x2, y2), (255, 255, 0), 2)
        elif label in ["sports ball", "ball"]:
            ball = (x1, y1, x2, y2)
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 0, 255), 2)

# اكتشاف العرقلة: إذا لاعبين قربوا من بعض كثير
def detect_fouls(players, threshold=60):
    fouls = []
    for i in range(len(players)):
        for j in range(i+1, len(players)):
            c1 = players[i][4]
            c2 = players[j][4]
            dist = np.linalg.norm(np.array(c1) - np.array(c2))
            if dist < threshold:
                fouls.append((players[i], players[j]))
    return fouls

# كشف العرقلة
fouls = detect_fouls(players)
for p1, p2 in fouls:
    x1, y1, x2, y2, c1 = p1
    x3, y3, x4, y4, c2 = p2

    # رسم سهم
    cv2.arrowedLine(image, c1, c2, (0, 0, 255), 3)
    cv2.putText(image, "Foul?", (min(x1, x3), min(y1, y3)-10), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

# حفظ الصورة المعدلة
cv2.imwrite(output_path, image)
print("تم حفظ الصورة المعدلة في:", output_path)
```

```python
import cv2
from ultralytics import YOLO

# تحميل الصورة
image_path = "keeper_outside_box.jpg"
image = cv2.imread(image_path)

# تحميل النموذج
model = YOLO("yolov8x.pt")  # تأكد أنه مدرب على اللاعبين والكرة

# كشف الأجسام
results = model(image)[0]

# إحداثيات منطقة الجزاء (حددها يدويًا حسب أبعاد الصورة)
penalty_area = [(150, 100), (450, 400)]  # [(x1, y1), (x2, y2)]

def is_inside_box(center, box):
    x, y = center
    (x1, y1), (x2, y2) = box
    return x1 <= x <= x2 and y1 <= y <= y2

keeper_center = None
ball_center = None

for box, cls in zip(results.boxes.xyxy, results.boxes.cls):
    x1, y1, x2, y2 = map(int, box)
    center = ((x1 + x2) // 2, (y1 + y2) // 2)
    label = model.names[int(cls)]

    if label in ['goalkeeper', 'person']:  # حسب نوع النموذج
        keeper_center = center
        cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(image, "Goalkeeper", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)

    elif label == 'sports ball':
        ball_center = center
        cv2.circle(image, ball_center, 5, (0, 0, 255), -1)
        cv2.putText(image, "Ball", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

# التحقق من الحالة
if keeper_center and ball_center:
    if not is_inside_box(keeper_center, penalty_area) and not is_inside_box(ball_center, penalty_area):
        cv2.putText(image, "خطأ: الحارس يمسك الكرة خارج منطقة الجزاء", (30, 60),
                    cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 3)

# رسم منطقة الجزاء
cv2.rectangle(image, penalty_area[0], penalty_area[1], (255, 0, 0), 2)

# حفظ النتيجة
cv2.imwrite("keeper_violation_result.jpg", image)
print("تم حفظ الصورة: keeper_violation_result.jpg")
```
